```
df_상권 = pd.read_csv('/gdrive/My Drive/data/상가업소정보_201912.csv',encoding='utf-8')
```

# 결측치 제거 

```
n = df_상권.isnull().sum()
not_use = n.sort_values(ascending=False).head(9)
df_상권.drop(not_use.index,axis = 1,inplace=True)
```

```
df_not_use = df_상권.columns[df_상권.columns.str.contains("코드|번호")]
df_상권.drop(df_not_use,axis =1,inplace=True )
df_상권.columns
```

#서울특별시 내 구별 노래방,나이트클럽 시설 


```
df_상권['상권업종중분류명'].unique()
```

```
df_노래방 =df_상권 [ (df_상권['상권업종중분류명'] == '무도/유흥/가무') &
          (df_상권['시도명'] == '서울특별시')]
df_노래방
```

```
df_구_노래방 = df_노래방.groupby(by='시군구명').count()
df_구_노래방
```

```
import json
geo = json.load(open("/gdrive/My Drive/data/서울_구별_경계.json"))
geo
```

```
import folium
g_map = folium.Map(location = [ 37.5502, 126.982],
                   tiles = "Stamen Terrain",
                   zoom_start = 11)
```

```
folium.Choropleth(geo_data=geo,
                  data= df_구_노래방,
                  columns=[df_구_노래방.index,"상호명"],
                  fill_color = "YlOrRd",
                  key_on = 'feature.properties.name'
                  ).add_to(g_map)
```
