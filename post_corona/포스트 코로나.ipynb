## 감소 후 증가

```
df_카드 = pd.read_csv("/gdrive/My Drive/2020_ai_project1/card_data/card_20200717.csv") #데이터 로드
```

```
df_카드
```

```
#필요없는 컬럼 삭제
df_카드.drop("adstrd_code", axis = 1, inplace = True)  
df_카드.drop("adstrd_nm", axis = 1, inplace = True) 
df_카드.drop("mrhst_induty_cl_code", axis = 1, inplace = True)
df_카드.drop("selng_cascnt", axis = 1, inplace = True)
```

```
sum_by_date_industry = df_카드["salamt"].groupby([df_카드["receipt_dttm"], df_카드["mrhst_induty_cl_nm"]]).sum()
sum_by_date_industry = sum_by_date_industry.reset_index() 
```

```
sum_by_date_industry.info()
```

```
for i, date in enumerate(sum_by_date_industry["receipt_dttm"]):

  if (20200100 <=  date <= 20200131) :
    sum_by_date_industry.loc[i,"month"] = "1월"
  elif (20200200 <=  date <= 20200229) :
    sum_by_date_industry.loc[i,"month"] = "2월"
  elif (20200300 <=  date <= 20200331) :
    sum_by_date_industry.loc[i,"month"] = "3월"
  elif (20200400 <=  date <= 20200430) :
    sum_by_date_industry.loc[i,"month"] = "4월"
  elif (20200500 <=  date <= 20200531) :
    sum_by_date_industry.loc[i,"month"] = "5월"
  else:
    sum_by_date_industry.loc[i,"month"] = "6월"
```

```
sum_by_date_industry.drop("receipt_dttm", axis = 1, inplace = True)
```

```
sum_by_date_industry
```

```
sum_by_date_industry_new = sum_by_date_industry["salamt"].groupby([sum_by_date_industry["month"], sum_by_date_industry["mrhst_induty_cl_nm"]]).sum()
sum_by_date_industry_new = sum_by_date_industry_new.reset_index() 
sum_by_date_industry_new
```

```
sum_by_date_industry_new["mrhst_induty_cl_nm"].unique()
```

감소 후 증가

```
#학원
df_학원 = sum_by_date_industry_new[sum_by_date_industry_new["mrhst_induty_cl_nm"].str.contains("학원")]
df_학원_월 = pd.DataFrame(df_학원["salamt"].groupby(df_학원["month"]).sum())
df_학원_월 = df_학원_월.add_suffix("_학원")
df_학원_월
```

```
#화장품
df_화장품 = sum_by_date_industry_new[sum_by_date_industry_new["mrhst_induty_cl_nm"].str.contains("화 장 품")]
df_화장품_월 = pd.DataFrame(df_화장품["salamt"].groupby(df_화장품["month"]).sum())
df_화장품_월 = df_화장품_월.add_suffix("_화장품")
df_화장품_월
```

```
#운동 헬스/레저
df_운동 = sum_by_date_industry_new[sum_by_date_industry_new["mrhst_induty_cl_nm"].str.contains("레져|헬스")]
df_운동_월 = pd.DataFrame(df_운동["salamt"].groupby(df_운동["month"]).sum())
df_운동_월 = df_운동_월.add_suffix("_운동")
df_운동_월
```

```
#음식점
df_음식점 = sum_by_date_industry_new[sum_by_date_industry_new["mrhst_induty_cl_nm"].str.contains("음식|한식|일식")]
df_음식점_월 = pd.DataFrame(df_음식점["salamt"].groupby(df_음식점["month"]).sum())
df_음식점_월 = df_음식점_월.add_suffix("_음식점")
df_음식점_월
```

```
#마트
df_마트 = sum_by_date_industry_new[sum_by_date_industry_new["mrhst_induty_cl_nm"].str.contains("대형할인점|슈퍼 마켓")]
df_마트_월 = pd.DataFrame(df_마트["salamt"].groupby(df_마트["month"]).sum())
df_마트_월 = df_마트_월.add_suffix("_마트")
df_마트_월
```

```
df_월 = pd.concat([df_학원_월, df_화장품_월, df_운동_월, df_음식점_월, df_마트_월], axis = 1)
```

```
df_월 = df_월.iloc[:5, :] #6월 데이터 삭제
df_월
```

```
fig_감소후증가= px.line(df_월, x = df_월.index, y = ["salamt_학원", "salamt_화장품", "salamt_운동"], title = '감소 후 증가')
py.iplot(fig_감소후증가)
```

```
fig_감소후증가2= px.line(df_월, x = df_월.index, y = "salamt_음식점", title = '감소 후 증가_음식점')
py.iplot(fig_감소후증가2)
```

```
                                                          fig_감소후증가3= px.line(df_월, x = df_월.index, y = "salamt_마트", title = '감소 후 증가_마트')
py.iplot(fig_감소후증가3)
```
