### 유흥주점 - 유동인구 비교

데이터 가져오기

```
df_서울_유동인구 = pd.read_csv("/gdrive/My Drive/2020_ai_project1/data/SeoulFloating.csv", encoding = "UTF-8") #데이터 로드
```

```
df_서울_유동인구.head()
```

```
df_서울_유동인구.info()
```

데이터 정제

* 필요 없는 행 삭제, 기준에 맞춰 필요한 데이터만 추출함

```
df_서울_유동인구.drop("province", axis = 1, inplace = True) # "province" 행 삭제 - 불필요
```

```
drop_70s = df_서울_유동인구[df_서울_유동인구["birth_year"] == 70].index #"birth_year"이 70인 행 삭제
df_서울_유동인구.drop(drop_70s, axis = 0, inplace = True)
```

```
df_서울_유동인구 = df_서울_유동인구[(df_서울_유동인구["hour"] >= 18) & (df_서울_유동인구["hour"] <= 23)] #시간대 18시 ~ 23시
```

```
df_서울_유동인구
```

```
#서울 내 유흥주점 개수 상위 5개 구 선택
condition = (df_서울_유동인구["city"] == "Gangnam-gu") | (df_서울_유동인구["city"] == 'Gwangjin-gu') | (df_서울_유동인구["city"] == 'Mapo-gu') | (df_서울_유동인구["city"] == 'Seocho-gu') | (df_서울_유동인구["city"] == 'Yeongdeungpo-gu')
df_서울_유동인구 = df_서울_유동인구[condition]
```

```
#코로나 발생 전 금토일 데이터 추출
condition_전 = (df_서울_유동인구["date"] == "2020-01-17") | (df_서울_유동인구["date"] == "2020-01-18") | (df_서울_유동인구["date"] == "2020-01-19")
df_서울_유동인구_전 = df_서울_유동인구[condition_전]
```

```
#코로나 발생 전 시간대별 유동인구 합계 계산
sum_by_date_hour_city_전 = df_서울_유동인구_전["fp_num"].groupby([df_서울_유동인구_전["date"], df_서울_유동인구_전["hour"], df_서울_유동인구_전["city"]]).sum()
sum_by_date_hour_city_전 = sum_by_date_hour_city_전.unstack().reset_index() #멀티인덱스 없애주기)

sum_by_date_hour_city_전["hour"] = sum_by_date_hour_city_전["hour"].astype(str) #문자열로 통일

#새로운 컬럼 생성
for i, date in enumerate(sum_by_date_hour_city_전["date"]):

  if date == '2020-01-17':
    sum_by_date_hour_city_전.loc[i,"date_hour"] = "금요일 " + sum_by_date_hour_city_전.loc[i,"hour"]
  elif date == '2020-01-18':
    sum_by_date_hour_city_전.loc[i,"date_hour"] = "토요일 " + sum_by_date_hour_city_전.loc[i,"hour"]
  else:
    sum_by_date_hour_city_전.loc[i,"date_hour"] = "일요일 " + sum_by_date_hour_city_전.loc[i,"hour"]

#필요 없는 컬럼 삭제
sum_by_date_hour_city_전.drop("date", axis = 1, inplace = True)
sum_by_date_hour_city_전.drop("hour", axis = 1, inplace = True)

sum_by_date_hour_city_전 = sum_by_date_hour_city_전.rename( #인덱스 변경
    {
     'Gangnam-gu': 'Gangnam-gu_전',
     'Gwangjin-gu': 'Gwangjin-gu_전',
     'Mapo-gu' : 'Mapo-gu_전',
     'date_hour' : "date_hour"
    }, axis = "columns"
)
```

```
#코로나 발생 후 - 고강도 사회적거리두기 기간 중 금토일 데이터 추출
condition_후 = (df_서울_유동인구["date"] == "2020-03-27") | (df_서울_유동인구["date"] == "2020-03-28") | (df_서울_유동인구["date"] == "2020-03-29")
df_서울_유동인구_후 = df_서울_유동인구[condition_후]
```

```
#코로나 발생 후 - 고강도 사회적거리두기 기간 중 시간대별 유동인구 합계 계산
sum_by_date_hour_city_후 = df_서울_유동인구_후["fp_num"].groupby([df_서울_유동인구_후["date"], df_서울_유동인구_후["hour"], df_서울_유동인구_후["city"]]).sum()

sum_by_date_hour_city_후 = sum_by_date_hour_city_후.unstack().reset_index() #멀티인덱스 없애주기

sum_by_date_hour_city_후["hour"] = sum_by_date_hour_city_후["hour"].astype(str) #문자열로 통일 후 
#새로운 컬럼 생성
for i, date in enumerate(sum_by_date_hour_city_후["date"]):

  if date == '2020-03-27':
    sum_by_date_hour_city_후.loc[i,"date_hour"] = "금요일 " + sum_by_date_hour_city_후.loc[i,"hour"]
  elif date == '2020-03-28':
    sum_by_date_hour_city_후.loc[i,"date_hour"] = "토요일 " + sum_by_date_hour_city_후.loc[i,"hour"]
  else:
    sum_by_date_hour_city_후.loc[i,"date_hour"] = "일요일 " + sum_by_date_hour_city_후.loc[i,"hour"]

#필요 없는 컬럼 삭제
sum_by_date_hour_city_후.drop("date", axis = 1, inplace = True)
sum_by_date_hour_city_후.drop("hour", axis = 1, inplace = True)

sum_by_date_hour_city_후 = sum_by_date_hour_city_후.rename( #인덱스 변경
    {
     'Gangnam-gu': 'Gangnam-gu_후',
     'Gwangjin-gu': 'Gwangjin-gu_후',
     'Mapo-gu' : 'Mapo-gu_후',
     'Seocho-gu' : 'Seocho-gu_후' ,
     'Yeongdeungpo-gu' : 'Yeongdeungpo-gu_후' , 
     'date_hour' : "date_hour"
    }, axis = "columns"
)
```
