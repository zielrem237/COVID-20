주요 5개국 코로나 바이러스 감염,사망,완치자 


```
# 5/30 날짜의 COVID-19 데이터 url
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/05-30-2020.csv"
```

```
# 데이터 불러오기
import pandas as pd 
df = pd.read_csv(url, encoding='utf-8')
df
```

```
df_world = df.groupby(['Country_Region'])['Confirmed','Deaths','Recovered'].sum()
df_world.sort_values(by='Confirmed',ascending=False,inplace=True)
df_world = df_world.head(5)
```
